```python
# creating a binary tree
class node:
  def __init__(self,val):
    self.val=val
    self.left=None
    self.right=None

root = node(1)
root.left = node(2)
root.right = node(3)
root.left.left = node(4)
root.left.right = node(5)


# inorder trevalser
def inorder(root):
  if root is None:
    return
  inorder(root.left)
  print(root.val,end=' ')
  inorder(root.right)
```

```python
# height of BT
def height(root):
  if root is None:
    return 0
  return max(height(root.left),height(root.right))+1

# diameter
def dia(self,root):
  self.ans=0
  def depth(node):
    if not node:
      return 0
    left = depth(node.left)
    right = depth(node.right)

    self.ans = max(self.ans,left+right)
    return max(left,right)+1

  depth(root)
  return self.ans
```

```python
print(height(root))
inorder(root)
```

```python
# print diameter
print(dia(root,root))
```
