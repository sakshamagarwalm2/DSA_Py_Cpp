```python
# creating a binary tree
class node:
  def __init__(self,val):
    self.val=val
    self.left=None
    self.right=None

root = node(1)
root.left = node(2)
root.right = node(3)
root.left.left = node(4)
```

```python
# insersion
def insert(root,key):
  if root is None:
    return node(key)
  queue = [root]
  while queue:
    temp = queue.pop(0)
    if temp.left is None:
      temp.left = node(key)
      break
    else:
      queue.append(temp.left)
    if temp.right is None:
      temp.right = node(key)
      break
    else:
      queue.append(temp.right)
  return root

# level
```
