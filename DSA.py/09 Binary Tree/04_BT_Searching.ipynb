```python
# creating a binary tree
class node:
  def __init__(self,val):
    self.val=val
    self.left=None
    self.right=None

root = node(1)
root.left = node(2)
root.right = node(3)
root.left.left = node(4)
root.left.right = node(5)


# inorder trevalser
def inorder(root):
  if root is None:
    return
  inorder(root.left)
  print(root.val,end=' ')
  inorder(root.right)
```

```python
# searching
def searchdfs(root,val):
  if root is None:
    return False
  if root.val==val:
    return True
  return searchdfs(root.left,val) or searchdfs(root.right,val)
```
