#Adjacency List Representation of Graph:
This graph is represented as a collection of linked lists. There is an array of pointer which points to the edges connected to that vertex.


image.png


```python
# For direction Graph --> 1
# For undirection Graph --> 0

def add_edge(adj,i,j,directed):
  # adj is list of lists: adj=[[]]
  # i and j are vertices
  if directed==0: # For undirected graph
    adj[i].append(j)
    adj[j].append(i)
  else: # fro directed graph
    adj[i].append(j)
```

```python
def display_graph(adj):
  for i in range(len(adj)):
    print(i,end=" --> ")
    for j in adj[i]:
      print(j,end=" ")
    print()
```

```python
Vertices = 5
adj = [[] for i in range(Vertices)] # Creating instance
```

```python
# For Undirectional Graph
add_edge(adj,0,1,0)
add_edge(adj,0,2,0)
add_edge(adj,0,3,0)
add_edge(adj,1,4,0)
add_edge(adj,2,3,0)
add_edge(adj,3,4,0)

display_graph(adj)
```

```python
#  For Directional graph
add_edge(adj,0,1,1)
add_edge(adj,0,2,1)
add_edge(adj,3,0,1)
add_edge(adj,1,4,1)
add_edge(adj,3,2,1)
add_edge(adj,4,3,1)

display_graph(adj)
```
